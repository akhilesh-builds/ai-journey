```python
#https://www.youtube.com/watch?v=vmEHCJofslg

import pandas as pd
df = pd.read_csv('pokemon_data.csv') #df stand for dataframe
#print(df)
print(df.head(3))#print the first 3 rows
print("\n\n")
print(df.tail(3)) #print the last 3 rows
print("\n\n")

#Read the data
print(df.columns) #print the columns
print("\n\n")

#print(df['Name']) #print the name column
print(df['Name'][0:5])
print("\n\n")

print(df[['Name','Type 1','HP']][0:5]) #print the name, type 1 and HP columns
print("\n\n")

print(df.iloc[0]) #print the first row
print("\n\n")
print(df.iloc[0:5]) #print the first 5 rows
print("\n\n")
print(df.iloc[0,1]) #print the first row and second column
print("\n\n")

#for index, row in df.iterrows():
    #print(index,row) #print the index and the row
    #print(index, row['Name']) #print the index and the name column only

df.loc[df['Type 1'] == 'Fire'] #print the rows where the type 1 is fire
df.describe() #print the statistics of the numerical columns
df.sort_values('Name', ascending=False) #sort the dataframe by name in descending order
df.sort_values(['Type 1','HP'], ascending=[1,0]) #sort the dataframe by type 1 in ascending order and HP in descending order

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] #create a new column called total which is the sum of all the stats
df.head(5) #print the first 5 rows to see the new column

df.drop(columns=['Total'], inplace=True) #drop the total column Same as df = df.drop(columns=['Total'])
df.head(5) #print the first 5 rows to see that the total column is

df['Total'] = df.iloc[:, 4:10].sum(axis=1) #create a new column called total which is the sum of all the stats using iloc. 
#Axis=1 means (horizontal) we want to sum across the columns. Axis=0 would mean(vertical)we want to sum down the rows.
df.head(5) #print the first 5 rows to see the new column

#to move the total column to the column after column speed
cols = list(df.columns)
cols.remove('Total')
speed_pos = cols.index('Speed')
cols.insert(speed_pos + 1, 'Total')
df = df[cols] #reorder the columns using the new column order
df.head(5) #print the first 5 rows to see the new column order

df.to_csv('modified_pokemon_data.csv',index=False) #export the dataframe to a new csv file without the index
```

The rule: .sort_values(), .head(), .filter(), and most pandas operations return new DataFrames. You should almost never reassign back to df unless you specifically want to permanently transform your dataset. Even then, prefer making the intent explicit

You overwrote df with the sorted+filtered result. That means your original 10-row DataFrame is now permanently a 3-row DataFrame. If you ran Exercise 4 next, your data would be gone.
This is a real pandas trap and worth locking now:
The mutation pattern (what you did):
pythondf = df.sort_values(by='Avg_Nightly_Rate', ascending=False).head(3)
This replaces df with the top-3 result. The original 10 hotels are no longer accessible.
The query pattern (what you wanted):
pythontop_3 = df.sort_values(by='Avg_Nightly_Rate', ascending=False).head(3)
print(top_3)
This creates a new variable top_3 for the result, and df stays untouched as your full dataset.

The rule: .sort_values(), .head(), .filter(), and most pandas operations return new DataFrames. You should almost never reassign back to df unless you specifically want to permanently transform your dataset. Even then, prefer making the intent explicit
