**What stuck:**
Conditional probability = how the probability of A changes once you know B 
is true. Notation: P(A|B), read "probability of A given B." Formula: 
P(A|B) = P(A and B) / P(B) — intuitively, "out of all the B cases, what 
fraction also has A."

**What was fuzzy:**
Concept is little fuzzy and not easy to grasp.anki

**Where I'll see this in AI:**
Every classifier output is P(class | input). LLM next-token generation is 
P(next_word | previous_words). Bayes' theorem (coming next) is built on 
flipping conditional probabilities. Hotel Concierge will use it for 
"given this guest asked X, what do they probably want."
