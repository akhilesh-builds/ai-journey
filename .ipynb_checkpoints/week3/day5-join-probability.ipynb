## Independent Events + Joint Probability (Fri May 29)

**What stuck:**
Two events are INDEPENDENT if knowing one tells you nothing about the other. 
For independent events: P(A and B) = P(A) × P(B). For ANY events (the general 
case): P(A and B) = P(A) × P(B|A) — the chain rule. The independence formula 
is just the chain rule's special case (because when independent, P(B|A) = P(B)).

The hotel example shows why this matters: under independence, expected 
P(long-stay AND room service) = 0.6 × 0.4 = 24%. Actual is 0.6 × 0.75 = 45%. 
The 21-point gap reveals positive correlation between the two behaviors — 
hidden information that the marginals alone don't show.

**What was fuzzy / new realization:**
Bayes' theorem isn't a separate concept — it's the chain rule applied twice 
and rearranged. P(A|B) = P(B|A) × P(A) / P(B). Will do reps next week.

**Where I'll see this in AI:**
Joint and conditional probabilities are the foundation of every probabilistic 
model — Naive Bayes classifiers (which literally assume features are independent, 
then use the multiplication rule), graphical models, every "what's the chance 
of X given Y" question. Hotel Concierge will use this to detect guest-behavior 
correlations and personalize recommendations.


Marginal probability P(A) Unconditional likelihood of A
Joint probability P(A and B) Likelihood of both happening
Independence rule P(A and B) = P(A) × P(B) Joint, when events are independent 
Chain rule P(A and B) = P(A) × P(B | A) Joint, for any events 
Conditional probability P(A | B) = P(A and B) / P(B) Likelihood of A given B happened 
Bayes' theorem P(A | B) = P(B | A) × P(A) / P(B) Flipping a conditional
