P(fraud | flagged) = true positives ÷ total flagged

Correction:
Strong work — you got the structure completely right, and you caught your own error mid-stream, which is the most important part. Let me grade precisely.

20 fraud, 980 legit. Correct.
Step 2 — True positives. This is where the slip happened.
You wrote "18 are legit" — but that's not legit bookings. That's 18 = the true positives (90% of 20 fraud caught = 18). You mislabeled it as "legit" but the number is right and you used it correctly later. So the math was right, the label was wrong. Quick fix: 18 is "real fraud that got caught," not "legit."
Step 3 — False positives. ✅

Step 4 — Total flagged. You caught your own bug here. 👏
You first wrote 20 + 40, then immediately questioned it: "should it be 18+40?" Yes — it should be 18 + 40. Here's why, and this is the subtle part worth locking:
The total flagged = everything the detector flagged = (real fraud it caught) + (legit it falsely flagged). The real fraud it caught is 18, not 20 — because the detector missed 2 of the 20 frauds (it only catches 90%). Those 2 missed frauds didn't get flagged, so they're not in the "flagged" pile.
So: total flagged = 18 + 40 = 58.
Step 5 — Final answer.
You wrote 18/60. The numerator (18) is right. The denominator should be 58, not 60.

P(fraud | flagged) = 18 / 58 ≈ 31%

Question 2: 
Base-Rate Problem 2 of 3
Different context — medical screening this time. Same pattern. Watch for the trap.

AI skin-cancer screening app. A dermatology startup builds an AI that screens photos of moles for melanoma. In the population using the app, 2 out of every 1,000 moles photographed are actually melanoma (it's rare). The AI correctly identifies 95% of real melanomas. It falsely flags 3% of harmless moles as suspicious.
A mole just got flagged by the AI as suspicious. What's the probability it's actually melanoma?

Same five steps:

How many moles are actually melanoma? How many are harmless?
Of the real melanomas, how many get caught (true positives)?
Of the harmless moles, how many get falsely flagged (false positives)?
Total flagged = true positives + false positives
P(melanoma | flagged) = true positives ÷ total flagged

Answer:
2 actual malenomas 998 harmless
true positives = 95% of 2 ~2
False positives = 3% of 998 ~30
p(malenoma|suspicious) = 2/32 = 1/16

A flagged mole is only ~6% likely to be melanoma. That sounds alarming — "the AI is wrong 94% of the time!" — but here's the nuance an AI EM has to understand: that's not a bad model. For a screening tool, you want high recall (catch 95% of real melanomas — don't miss cancer) even at the cost of low precision (lots of false alarms). The false positives go to a dermatologist for a 30-second visual check. The cost of a false positive (a quick doctor visit) is tiny compared to the cost of a false negative (missed cancer). The base rate forces a deliberate precision-recall tradeoff, and the right tradeoff depends on the real-world cost of each error type.
Contrast with the fraud case: there, auto-cancelling a flagged booking hurts the 69% legit customers, so you route to human review instead. Different costs → different product decision, same underlying base-rate math.
That reasoning — "what does each type of error actually cost, and how should that shape the system?" — is exactly what gets asked in AI EM interviews. You're touching it now, weeks ahead of schedule.

My Question: 
Can you explain again why lower base rate makes the lower precision?

Good question to ask — and the fact that you want the why behind the pattern, not just the pattern, is exactly right. Let me build the intuition so it's not just "the table said so."
The core idea: flagged results come from two pools, and they grow at different rates.
Every flag the classifier raises comes from one of two places:

True positives — real cases it correctly caught. These come from the small "actual" population.
False positives — harmless cases it wrongly flagged. These come from the large "harmless" population.

Precision is the ratio: true positives ÷ (true positives + false positives). So precision is really asking: "of everything I flagged, how much came from the real pool vs. the false-alarm pool?"
Now here's why the base rate controls it.
Watch what happens to each pool as the base rate drops. Let's hold the classifier's quality fixed (say 90% catch rate, 3% false-positive rate) and only shrink the base rate:
High base rate — 200 in 1,000 are real:

True positives: 90% of 200 = 180
Harmless population: 800 → false positives: 3% of 800 = 24
Precision = 180 / (180+24) = 88%

Medium base rate — 20 in 1,000 are real:

True positives: 90% of 20 = 18
Harmless population: 980 → false positives: 3% of 980 = 29
Precision = 18 / (18+29) = 38%

Low base rate — 2 in 1,000 are real:

True positives: 90% of 2 = 1.8
Harmless population: 998 → false positives: 3% of 998 = 30
Precision = 1.8 / (1.8+30) = 6%

See what moved and what didn't?
The false positives barely changed — 24, 29, 30. They're roughly stuck around 30, because they come from the harmless pool, which is always ~1,000-ish regardless of base rate.
The true positives collapsed — 180 → 18 → 2. They shrank in direct proportion to the base rate, because they come from the real pool, which is what the base rate is shrinking.
So as the base rate drops, your numerator (true positives) shrinks fast while your false-positive count stays roughly fixed. A shrinking numerator over a roughly-fixed denominator → precision falls. The false alarms swamp the dwindling real cases.
The one-sentence intuition to lock:

When the real thing is rare, even a small false-positive rate applied to the huge "everything else" population produces more false alarms than there are real cases — so most of your flags are false alarms.

A mental image that sticks: imagine searching a haystack for needles. If there are 200 needles in the hay, your metal detector's occasional false beeps don't matter much — you find lots of real needles. If there are only 2 needles in the same giant haystack, the same detector with the same occasional false beeps will beep mostly at hay, because there's so much more hay than needle. The detector didn't get worse. The needles just got rarer. Rarity, not detector quality, drove the precision down.
That's why a 95%-accurate cancer test gives a positive that's only 6% likely to be real — the disease is the rare needle, and the healthy population is the enormous haystack.
