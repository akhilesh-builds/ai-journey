**What stuck:**
The dot product of two vectors produces a single number that measures how 
aligned they are. Geometrically: project one onto the other, multiply by 
the second's length. Positive = same direction, zero = perpendicular, 
negative = opposite. Mechanically: multiply matching components, add.

**What was fuzzy:**
The duality part — the deep correspondence between vectors and 1×N matrices 
(linear transformations). Need a second watch tomorrow morning fresh.

**Where I'll see this in AI:**
Three places: (1) Cosine similarity for matching embeddings — "how similar 
are these two sentences." (2) Attention in transformers — query · key 
literally is the attention score. (3) Every neural network neuron — output 
is dot(weights, inputs) + bias. Dot products are the atomic operation of 
deep learning.
