## Episode 1 ‚Äî Vectors: What Even Are They?

### The Three Views of a Vector
Same object, three perspectives. Fluency means moving between them.

1. **Physics view** ‚Äî a vector is an **arrow in space** with a magnitude (length) and a direction. It can live anywhere; only length + direction matter.
2. **CS / Data view** ‚Äî a vector is an **ordered list of numbers**: `[1, 2]` is a 2D vector. Position of each number matters.
3. **Mathematician's view** ‚Äî a vector is **anything that can be added to another vector and scaled by a number** (abstract: functions and polynomials qualify too).

**Core message:** linear algebra is the bridge ‚Äî being able to see `[1, 2]` as BOTH a list of numbers AND an arrow in 2D space is a superpower.

### Vector Addition (Tip-to-Tail)
To add `(2, 3) + (‚àí1, 4)`:
- Draw vector `(2, 3)` from the origin ‚Üí tip lands at `(2, 3)`.
- Start the next vector `(‚àí1, 4)` FROM that tip (not from origin).
- Final position: `(1, 7)` ‚Üí that's the sum.

This is WHY coordinate-wise addition works: `x1 + x2` and `y1 + y2`.

### Scalar Multiplication
Multiplying a vector by a scalar does TWO things:
- **Sign flips direction** (negative scalar reverses the arrow 180¬∞)
- **Magnitude stretches or shrinks length** (|scalar| > 1 stretches, < 1 shrinks)

Example: `(3, ‚àí2) √ó (‚àí1.5) = (‚àí4.5, 3)` ‚Üí flipped direction AND 1.5√ó longer.

### Column Notation
Vectors are written vertically (as columns) because matrices later "act on" vectors via matrix-vector multiplication ‚Äî and the shape rule requires the vector be a column. Foreshadows Ep 3.

## Episode 2 ‚Äî Linear Combinations, Span, and Basis Vectors

### Linear Combination
An **operation** ‚Äî take two (or more) vectors, scale each by a number, add them:

    a ¬∑ v‚Üí + b ¬∑ w‚Üí

This produces a single vector. Every choice of `a, b` gives a different result.

### Span
The **set of ALL vectors reachable** by every possible linear combination of a given set of vectors. It's the geometric shape those vectors can "cover."

| Setup                                          | Span                |
|-----------------------------------------------|---------------------|
| 2 linearly independent vectors in 2D          | The whole 2D plane  |
| 2 linearly dependent vectors in 2D            | Just a line         |
| 3 linearly independent vectors in 3D          | All of 3D space     |
| 3rd vector lies on plane of first two (3D)    | Just that plane     |

### East-North-Up Analogy (3D span intuition)
- v‚Üí = east, w‚Üí = north ‚Üí they span the **floor plane**.
- Add u‚Üí pointing **up** ‚Üí span becomes **all of 3D**.
- Add u‚Üí pointing **northeast** (still on floor) ‚Üí no new direction ‚Üí span stays the floor. u‚Üí was **redundant**.

### Linear Dependence vs. Independence
- **Linearly dependent:** one vector is a **scalar multiple** of another ‚Üí `u‚Üí = c ¬∑ v‚Üí` for some scalar `c`.
  - Geometric test: they lie on the **same line through the origin** (parallel, same or opposite direction).
- **Linearly independent:** no scalar makes them equal ‚Üí they point in **genuinely different directions**.

**Examples:**
- `(2, 4)` and `(1, 2)` ‚Üí dependent (2 ¬∑ (1,2) = (2,4))
- `(‚àí3, ‚àí6)` and `(1, 2)` ‚Üí dependent (‚àí3 ¬∑ (1,2) = (‚àí3,‚àí6))
- `(1, 2)` and `(3, 1)` ‚Üí independent

### Basis
A **basis** of a vector space is a set of **linearly independent vectors that span the whole space**.

- 2D ‚Üí basis has exactly **2** vectors
- 3D ‚Üí basis has exactly **3** vectors
- n-D ‚Üí basis has exactly **n** vectors

The **dimension** of a space is literally the number of vectors in its basis.

**Standard basis in 2D:** √Æ = `(1, 0)`, ƒµ = `(0, 1)` ‚Äî the default, but not the only one. Any two linearly independent 2D vectors form a valid basis.

### Consequence (Theorem)
In 3D, if you have 4 vectors, **at least one MUST be redundant** (dependent). You can't have 4 independent vectors in 3D ‚Äî there's no room.

---

## Key Corrections I Needed
- **Linear combination = the operation.** **Span = the set of all results.** Different levels of abstraction.
- **Dependence test:** `u‚Üí = c ¬∑ v‚Üí` for ANY scalar `c`, not just `c = 1`.
- **3rd vector expands span** only if it has ANY component perpendicular to the plane of the first two ‚Äî doesn't need to be pure z-axis.

## Preview for Ep 3
Every linear transformation (rotating, stretching, shearing, projecting) is essentially a **change of basis**. A matrix's columns describe where the standard basis vectors `√Æ, ƒµ` land after the transformation. Matrix √ó vector = reinterpreting that vector in a new basis.

# Ep4
If we do two linear transformation say first by rotating by 90 degrees counter clockwise and then apply shear. This is same as a liner transformation which is "Composition" of rotation and shear (transformation).
Suppose after composition the basis vectors changed to 
i-hat = 1,1
j-hat = -1,0

This is capturing both the transformation like rotation and shear rather than two successive ones. We can achieve the same through multiplication of two successive transformations:

Suppose we want to get transformation of a vector based on composition of rotation and shear
[x]
[y]

[1 1] ([0 -1][x]) = [1 -1][x]
[0 1] ([1. 0][y]) = [1  0][y]
shear.  Rotation
Here its going from right to left, meaning fir rotation is applied and then shear.

That means basis vectors
[1 1] [0 -1] = [1 -1]
[0 1] [1. 0] = [1  0]
shear.  Rotation = Composition

Read right to left. A¬∑B¬∑v = "transform v by B, then transform the result by A."

Why AB ‚âÝ BA: rotation-then-shear gives a different final shape than shear-then-rotation. Order matters, just like putting on socks before shoes vs. shoes before socks.

How you compute it: take each column of B (where B sends √Æ and ƒµ), then apply A to those landing spots. The result is the columns of the new matrix.

matrix multiplication = function composition (first do transformation B, then transformation A, written A¬∑B). It explains why matrix multiplication is not commutative (AB ‚âÝ BA) in a way that will finally make sense ‚Äî rotation-then-shear is different from shear-then-rotation.

Matrix multiplication is associative

(AB)C = A(BC)


```python
import numpy as np

# Two transformations
A = np.array([[2, 1], [0, 3]])      # a shear + scale
B = np.array([[0, -1], [1, 0]])     # 90¬∞ counterclockwise rotation

# Apply B first, then A  (this is what A @ B means ‚Äî right to left!)
AB = A @ B
BA = B @ A

print("A @ B (B first, then A):")
print(AB)
print()
print("B @ A (A first, then B):")
print(BA)
print()
print("Are they equal?", np.array_equal(AB, BA))
```
