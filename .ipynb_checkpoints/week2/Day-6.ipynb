Indexing and slicking

Acessing an element using the index notation is called idexing. 
arr[index]



```python
import numpy as np

arr = np.arange(10)
print(arr)

print(arr[3])

print(arr[0:4])

print(arr[:10:2])

arr2 = np.array([[2,3,4,5], [3,4,5,6], [4,5,6,7]])
print(arr2)
print(arr2.shape)
print(arr2.dtype)

print(arr2[2][3])

arr2[1:3, 1:3] = 0
print(arr2)

arr2 > 1
print(arr2 > 1)

arr[arr>1]
print(arr[arr>1])


```
