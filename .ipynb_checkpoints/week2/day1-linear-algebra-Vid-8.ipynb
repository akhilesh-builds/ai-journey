    Liek a matrix with 3 rows and 2 columns basically saying.
    1. two columns says there are 2 basis vectors in this transformation
    2. 3 row says that the basis vecots are described by 3 coordinates in this transformation
    therefore after this transformation a 2-d space (which is 2 basis vecots) will be transformed into 3-d space where the coordinates are represented by 3 coordinates. 
    So this is 2-d to 3-d transformation where the resultant transformation make s vector representd in 3-d space

    Same for a 2*3 column vector transformation:
    1. 3 columns represents 3 basis vectors
    2. 2 rows represents that each basis vector after this transformation will be represented in 2 coordinates. 
    This means that after this transformation a 3-d vector will tansform into 2-3 vector.
