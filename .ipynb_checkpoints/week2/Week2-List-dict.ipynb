File:
File operations usually follows 3 steps: Open, Use, Close
Open the file by telling the OS about the path

Use to either perform read or write operations

Close to flush out from memory and releasing the lock to avoid memory leaks. 

To open a file we use with keyword: 
Using this keyword makes the file auto close once the operation (code block under the open statement is done). It also makes sure to close the file even if the operation generated a crash.





```python
#Syntax
with open("myfile.txt", "r") as f:
    content = f.read()

#open() takes two arguments: the path and the mode.
#"r" is the mode — read. (Others: "w" write, "a" append. More on this in a sec.)
#as f — f is the variable name for the file handle inside the block. Convention is f but you can call it anything.
```

there are 4 modes of file open:
1. "r": read only. If file foesnt exist it raises an exception fileNotFound. If exists then it opens it for reading.
2. "w" write: if file doesnt exist then it creates a file and write to it. If file exists then it ERASES THE CONTENT AND CREATE FROM START.
3. "a" append: if file doesnt exist then it creates a file and write to it. If file exists then it opens it and write to end.
4. "r+" read+write: If file doesnt exist then it crashes.

Three read operations: 
1. read(): this fetches the file data as a one string. Loads all the data upfront and not good for large files. Eager loading. Gives one big string
2. readlines(): this fethes data and gives a list of lines of the files. Each line becomes an entry in the list. Eager load. Not good for large files
3. for line in f: this is lazy loading and reads one line as a time on demand. it doenst load full data at once. Good for large files. 

Gotchas:
1. Every line we read keeps the \n at the end. Prints fine but string comparison fails. "hello" != "hello\n"
use strip() to strip new line.
for line in f:
    clean = line.strip()
2. Once the file is red and cursor is at the end. If you want to read it again then use seek(0) funciton on the file. f.seek(0)


Write: 
with open("file.txt", "w") as f:
    f.write("yo")

with open("file.txt", "w") as f:
    f.writelines(["hello", "hi"])

Write doenst use new line unless specified
with open("file.txt", "w") as f:
    f.write("yo\n")

with open("file.txt", "w") as f:
    f.writelines(["hello\n", "hi\n"])

For a list we can also use sting method idiom
hotels = ["Wynn", "MGM", "Bellagio"]
f.write("\n".join(hotels))
