NumPy is multidimenstional array library. You can store 1d, 2d, 3d.... array using it. 
Why NumPy over list:
1. NumPy is faster than List
    1. NumPy uses Fixed type. Suppose we have numeric matrix and one of the value in there is 5. Machine sees 5 as binary.
        5 -> 00000101 (8 Bit = 1Byte) 
        In NumpY (int 32) = 00000000 00000000 00000000 00000101
        In NumpY (int 16) = 00000000 00000101
        In NumpY (int 8) = 00000101

        In List we have lot more info to store as int. Built in int type consist of various values
        Object Type. => 00000000 00000000 00000000 00000101 00000000 00000000 00000000 00000101
        Object Value => 00000000 00000000 00000000 00000101 00000000 00000000 00000000 00000101
        Reference Count => 00000000 00000000 00000000 00000101 00000000 00000000 00000000 00000101
        Size => 00000000 00000000 00000000 00000101

        So it requires a lot more space than NumPy. So computer reads less bytes quicker
    2. Also while Iterating NumPy doesnt do type check since all the values are of same type unlike list
    3. NumPy utilizes continuous memory. Unlike list will store pointers to scattered memory blocks which makes it slow as it has to get the values based on pointers. In NumPy, say array of 8 1d array will be stores in continuous memory. Its faster because of SIMD vector(Single instruction multiple data). Its like if you want to add all the values it will not do one at a time rather do all at once.
    Also it utilizes effective cache since data is in continuouse blocks.

2. Lists we can do insertion, deletion, appending, concatenate etc and NumPy has more
    
3. NumPy
    a = [1,2,3]
    b = [4,5,6]
    in list a*b is not allowed

    a = np.array([1,2,3])
    b = np.array([4,5,6])

    a * b is allowed in NumPy
4. Best for machine learning since many library build using NumPy
