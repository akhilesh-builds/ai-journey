## Week 5 Day 0 — What Machine Learning Is (Sun Jun 7)

**What stuck:**
ML is programming-by-example. Instead of writing if/else rules, you give 
the computer labeled examples (features + correct labels) and let it find 
the pattern. Function shape: features in → label out. Training learns the 
function, test data verifies it actually generalized instead of memorizing 
(overfitting). Two branches: supervised (have labels) and unsupervised 
(no labels). Week 5 is almost entirely supervised — the foundation for 
all downstream ML.

**The big picture for my career:**
Classical ML isn't what AI EM interviews directly test, but the vocabulary 
and frame are what every downstream conversation (deep learning, LLMs, 
RAG, evals) depends on. Installing the conceptual scaffold before adding 
words.

**Where I'll see this in AI:**
Every Anthropic/OpenAI/Google AI model is supervised learning applied at 
scale. Same skeleton: features in, label out, train, test. The capstones 
(Hotel Concierge + Tester Agent) will use the same frame — different 
features (text), different labels, same shape.
