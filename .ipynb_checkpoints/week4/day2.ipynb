First, the small terminology cleanup so the words don't get in the way:

Mean = the average. Add all the values, divide by how many. Same word you've used your whole life.
Median = the middle value when you sort everything. Half the data is below, half above.
Standard deviation (σ) = "how spread out is the data on average?" Small σ = tight cluster. Big σ = wide spread.

I haven't actually introduced "median" yet today — that's a curriculum word that'll come up. Don't worry about it for now. Today we only need: mean (average), standard deviation (spread), and "shape."
Now — let me show you the picture of what a distribution is. One image per concept.

Picture 1 — Uniform distribution (the dice roll case).
Imagine you roll a fair die 10,000 times and count how many times each face came up. If you plot it:

X-axis: the value you rolled (1, 2, 3, 4, 5, 6)
Y-axis: how often you got that value

You'd get six bars all roughly the same height. Every face shows up about 1/6 of the time.
That flat shape — every outcome equally likely — is called a uniform distribution. "Uniform" because the heights are uniformly the same.
The chart is the distribution. The shape is the information. A uniform distribution tells you "no value is more likely than any other."


Picture 2 — Normal distribution (the famous bell curve).
Now imagine you measure the height of 10,000 random adults and plot it the same way:

X-axis: height (in inches, say)
Y-axis: how many people had that height

You wouldn't get flat bars. You'd get a hill shape — peaked in the middle, tapering off on both sides. Most people cluster around the average. Very few are extremely short or extremely tall.
That hill shape is called the normal distribution or bell curve. It's the most common shape in nature.
Two numbers fully describe it:

μ (mean) — where the peak sits on the x-axis. The center of the hill.
σ (standard deviation) — how wide the hill is. Tall narrow hill = small σ. Short wide hill = large σ.

The picture you want in your head when I say "normal distribution" is a smooth bell-shaped hill. Mean is the center. Standard deviation is the width.
The vertical lines you see on bell curve images at "1σ, 2σ, 3σ" from the mean are the 68-95-99.7 rule drawn in:

68% of the data falls within one standard deviation of the mean (the central region)
95% falls within two standard deviations (a bit wider)
99.7% falls within three standard deviations (almost everything)

So if hotel nightly rates have mean $300 and σ = $50:

68% of nights: between $250 and $350 (one σ on each side of $300)
95% of nights: between $200 and $400 (two σ on each side)
99.7% of nights: between $150 and $450 (three σ on each side)

A $475 night sits more than 3σ above the mean. Less than 0.15% likely. That's how outlier detection works: anything past 2-3σ from the mean is suspiciously far from "typical" and gets flagged.

Picture 3 — Binomial distribution (count of successes).
Now a different setup. Flip a coin 10 times. Count how many heads you got. That count can be 0, 1, 2, 3 … all the way to 10.
Now repeat this experiment thousands of times — flip 10 coins, count heads, write down the count. Plot how often each count occurred:

X-axis: number of heads (0 through 10)
Y-axis: how often you got that count

You'd get a hill-shaped bar chart with the peak at 5 (which makes sense — 5 out of 10 is the expected number with a fair coin). 0 heads or 10 heads would be rare bars at the edges.
That's a binomial distribution. It's the distribution of "how many successes in n trials."
Two parameters:

n — total number of trials (in our example, 10 flips)
p — probability of success per trial (0.5 for a fair coin)

When p = 0.5, the binomial is symmetric like the picture above. When p is far from 0.5 (say, p = 0.9), the hill tilts toward the high end — most outcomes are near 9 successes.
Why this matters in AI: every classifier accuracy measurement is binomial. "Model got 850 out of 1000 examples right" → that 850 came from a binomial distribution with n=1000 and p=true_accuracy. The spread of that distribution tells you how uncertain your accuracy estimate is.
