```python
import pandas as pd
import numpy as np

# Synthetic but realistic hotel booking data
np.random.seed(42)

n = 200

hotels = pd.DataFrame({
    'BookingID': range(1, n + 1),
    'Nights': np.random.choice([1, 2, 3, 4, 5, 7, 10, 14], size=n,
                                p=[0.30, 0.25, 0.20, 0.10, 0.08, 0.04, 0.02, 0.01]),
    'Nightly_Rate': np.round(np.random.normal(loc=300, scale=80, size=n), 2),
    'Guests': np.random.choice([1, 2, 3, 4], size=n, p=[0.20, 0.55, 0.15, 0.10]),
    'Booking_Channel': np.random.choice(['Direct', 'OTA', 'Corporate', 'Travel Agent'],
                                          size=n, p=[0.35, 0.45, 0.15, 0.05]),
    'Is_Repeat_Customer': np.random.choice([True, False], size=n, p=[0.25, 0.75])
})

# Clip nightly rate to realistic floor (no negative rates)
hotels['Nightly_Rate'] = hotels['Nightly_Rate'].clip(lower=80)

# Total revenue per booking
hotels['Total_Revenue'] = hotels['Nights'] * hotels['Nightly_Rate']

print(hotels.groupby('Booking_Channel')["Nightly_Rate"].mean())

print(hotels.head(10))
print(f"\nShape: {hotels.shape}")
print(hotels['Nightly_Rate'].mean())
print(hotels['Nightly_Rate'].std())
print(hotels['Nightly_Rate'].min())
print(hotels['Nightly_Rate'].max())


#calculate the how many rows has nightly rate between $228.69 and $382.49
lower = hotels['Nightly_Rate'].mean() - hotels['Nightly_Rate'].std()
upper = hotels['Nightly_Rate'].mean() + hotels['Nightly_Rate'].std()
count_nightly_rate_range = hotels[(hotels['Nightly_Rate'] >= lower) & (hotels['Nightly_Rate'] <= upper)].shape[0]
percentage_nightly_rate_range = (count_nightly_rate_range / hotels.shape[0]) * 100
print(f"\nNumber of bookings with nightly rate between ${lower:.2f} and ${upper:.2f}: {count_nightly_rate_range} ({percentage_nightly_rate_range:.2f}%)")


print(hotels['Total_Revenue'].mean())
print(hotels['Total_Revenue'].std())
print(hotels['Total_Revenue'].min())
print(hotels['Total_Revenue'].max())
print(hotels['Total_Revenue'].median())
print((hotels['Total_Revenue'] > 2500).sum() / len(hotels) * 100)
print(hotels.groupby('Booking_Channel')['Total_Revenue'].median())

print(hotels.groupby("Booking_Channel").size())
print(hotels[hotels["Booking_Channel"] == "OTA"]["Total_Revenue"].mean())
print(hotels.groupby("Booking_Channel")["Total_Revenue"].mean())
print(hotels.groupby("Booking_Channel")["Total_Revenue"].mean().sort_values(ascending=False))
```
