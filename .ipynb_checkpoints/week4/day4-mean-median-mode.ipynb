## Skewness + Median Lock-In (Thu Jun 4 — fresh-brain retry)

**What stuck:**
A distribution is *skewed* when it's asymmetric — one side has a longer tail 
than the other. Right-skewed = long tail to the right (income, revenue, 
latency). Left-skewed = long tail to the left (extended-stay bookings, 
end-of-quarter sales).

Three measures of "center" behave differently in skewed data:
- Mode = the peak (most common value)
- Median = the middle value when sorted (position-based, ignores tail magnitude)
- Mean = the average (magnitude-based, gets pulled toward the tail)

For right-skewed: Mode < Median < Mean.
For left-skewed: Mean < Median < Mode.
For symmetric (normal): Mean = Median = Mode.

For my hotel data: Total_Revenue mean = $810, median = $654. The $156 gap is 
the skew signature — the long tail of high-revenue long-stays pulls the mean 
above the typical (median) booking.

**What was fuzzy yesterday, lock today:**
The mean isn't *wrong*, it's just measuring a different thing than "typical." 
Median is robust against outliers because it's position-based. Mean is sensitive 
to outliers because it's magnitude-based. Use median for "typical," mean for 
aggregates/totals, both with the gap explained when reporting to stakeholders.

**The business framing (own thinking):**
The mean-median gap is itself a strategy lever. A high mean with low median 
means most revenue comes from a small set of long stays — closing the gap 
would mean shifting the booking mix toward longer stays via marketing, packages,
loyalty programs. The statistical concept becomes a product strategy question.

**Where I'll see this in AI:**
Every business metric I'll be asked about as an AI EM is right-skewed: 
revenue, latency, engagement, model inference time, transaction sizes. 
Reporting mean alone on these is misleading and senior interviewers will 
ask about it. Always report mean + median + gap interpretation for skewed data.
P50 (median) latency vs P99 latency tell completely different stories about 
system performance.

**Mean vs. Median, sharpened (lock):**
- Mean × Count = Total. This is mean's superpower — connects individual 
  values to aggregates.
- Median has NO connection to total. It's a position, not a magnitude.
- Critical: the mean you use must match the scope of the question. 
  Overall mean ≠ short-stay mean. Use group-specific means for group-specific 
  questions, or just .sum() directly when you have all the data.
- Three tools, not two: mean for forecasting/extrapolating, median for 
  "typical," .sum() for exact known totals.
